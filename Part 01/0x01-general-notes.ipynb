 # Concepts

 **CHAIN RULE** It how we can determine how much the input nodes effect the result of the output nodes. It is the derivative of the output with respect to the input. It is the sum of the derivatives of the output with respect to each of the intermediate nodes multiplied by the derivative of the intermediate node with respect to the input.

 **DELTA RULE** defines how the neurones are updated. It is the derivative of the error with respect to the output multiplied by the derivative of the output with respect to the input.

 diffrence between chain rule and delta rule is that the chain rule is used to determine how much the input nodes effect the result of the output nodes. The delta rule is used to determine how much the output nodes effect the error.

 **BACKPROPOGATION** backpropagation is the whole suite of algorithms that are used to calculate the derivatives of the output with respect to the input. It is the process of calculating the gradient of the loss function with respect to the weights of the neural network.

**GRADIENT DESCENT** Gradient descent is the process of updating the weights of the neural network in the direction of the negative gradient of the loss function. This does not define how the gradient is calculated, only how the weights are updated.

**Stochastic** random distrubution that can be predicted but not preciesly calculated.

**SCALERS vs TENSOR** A scaler is a single number. A tensor is a multi-dimensional array of numbers. It says that we just single values to represent the weights and biases of the neural network. We can use a tensor to represent the weights and biases of the neural network. This is the data the neural net is training on. Tensors are just arrays of scalers


```python
from micrograd.engine import Value
```

```python
a = Value(-4.0)
b = Value(2.0)

c = a + b
d = a * b + b**3
c += c + 1
c += 1 + c + (-a)
d += d * 2 + (b + a).relu()
d += 3 * d + (b - a).relu()
e = c - d

f = e**2
g = f / 2.0
g += 10.0 / f

print(f'{g.data:.4f}')
g.backward()
print(f'{a.grad:.4f} {b.grad:.4f}')
```
